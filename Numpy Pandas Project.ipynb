##### NUMPY

```python
import numpy as np
```

1. Konstruera en array med 10 kolumner och 5 rader av float-tal. Inget tal får vara
samma

```python
random_array1 = np.random.choice(np.arange(1, 100), replace = False, size=(5, 10))   
random_array1 = random_array1.astype(float)

random_array1
```

2. Konstruera en ny array till med 10 kolumner och 7 rader av float-tal. Den skall inte
vara samma som ovan array..

```python
random_array2 = np.random.choice(np.arange(1, 100), replace = False, size=(7, 10))   
random_array2 = random_array2.astype(float)

random_array2
```

3. Gör om de ovan arrayerna så att de har 5 kolumner men innehåller samma data.

```python
random_array1_reshaped = random_array1.reshape(10, 5)
random_array1_reshaped
```

```python
random_array2_reshaped = random_array2.reshape(14, 5)
random_array2_reshaped
```

4. Slå ihop de bägge arrayerna till en array.

```python
merged_array = np.vstack([random_array1_reshaped,random_array2_reshaped])
merged_array
```

5. Ändra ordningen på kolumnerna så att de har ordningen kolumn5, kolumn3, kolumn4, kolumn1 och kolumn2.

```python
merged_array[:, [4, 2, 3, 0, 1]]
```

10. Skapa en ny dataframe med 10 rader. Den innehåller de 10 rader med lägsta antalet överfall
per invånare (Rate_per_10000_population), men mer än 5000 invånare. 

```python
lowest_assults = assults[assults['Population_mid_point_2015'] > 5000]
df_sorted = lowest_assults.sort_values(by=['Rate_per_10000_population'], ascending=True)
top_ten_df_sorted = df_sorted[:10]
top_ten_df_sorted
```

11. Skapa en ny dataframe ”regions” som innehåller sammansatt data per region
(Region_2013_label). Lägg till en ny kolumn som innehåller ”Rate_per_10000_population”
fast för hela regionen.

```python
regions = df.groupby("Region_2013_label").sum()
regions["Rate_per_10000_population"] = regions["Victimisations_calendar_year_2015"] / regions["Population_mid_point_2015"] * 10000
regions["Rate_per_10000_population"] = regions["Rate_per_10000_population"].astype(int)
regions[["Victimisations_calendar_year_2015", "Population_mid_point_2015", "Rate_per_10000_population"]]
```

12. Lägg till en ny kolumn i dataframen med regionerna som visar i % hur många av alla överfall
som sker i den regionen

```python
regions["region_crime(%)"] = regions["Victimisations_calendar_year_2015"] / regions["Victimisations_calendar_year_2015"].sum() * 100
regions[["Victimisations_calendar_year_2015", "region_crime(%)"]]
```
